```python
import numpy as np
import torch
import random
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img_new = Image.open("...\\name.jpg")
```

```python
from torchvision import transforms
import torchvision
```

### Оригинальное изображение (original image).

```python
# Original image
img_new
```

###                                                            1 Обрезать изображение (Cut image)

```python
# Cut the image
transfomation = transforms.CenterCrop(400)
transfomation(img_new)
```

### 2. Случайным образом изменить яркость, контраст и насыщенность. (Randomly change the contrast, brightnees and/or saturation)

```python
transforms.ColorJitter(100, 100, 42)(img_new)
```

### 3. Вырежит 5 кусков из фото. 4 по углам, один в центре (Cut 5 pieces, 4 in the corner, 1 in the center)

```python
arr = transforms.FiveCrop(200)(img_new)

for el in range(5):
    plt.imshow(arr[el])
    plt.show()
```

### 3. Конвертирует изображение в оттенки серого (Convert image to gray)

```python
transforms.Grayscale(3)(img_new)
```

### 4. Дополняет изображеие пикселями. Для каждой стороны можно установить свою длину отступа. Также можно задать цвет. (Padding)

```python
transforms.Pad(padding=(100, 50, 20, 11), fill=123)(img_new)
```

### 5. Может повернуть изображение, отобразить только часть относительно центра, отмасштабировать. (Rotation, translat, scaling)

```python
transforms.RandomAffine(40, translate=[0.5, 0.5], scale=[.5, 3])(img_new)
```

### 6. Случайным образом применит или не применит к изображению трансформации из передаваемого списка. Есть возможность настроить вероятность. (Apply randomly a list of transformations).
То есть, в качестве параметров метод принимает список трансформациий и вероянтность (по умолчанию  задано 0.5). В случае
применения метод применит к изображению все трансформации из списка.

```python
transforms.RandomApply([transforms.Pad(padding=(100, 50, 20, 11), fill=123), transforms.Grayscale(3)], p=.5)(img_new)
```

### 7. Случайным образмо обязательно применит к изображениию одну траснформацию из списка. На выходе никогда не будет изначальное изображение. (Randomly apply on transformation from list.)

```python
transforms.RandomChoice([transforms.Pad(padding=(100, 50, 20, 11), fill=123), transforms.Grayscale(3)])(img_new)
```

### 8. Случайным образом вырезает заданый размер изображения. (Randomly crop some piece of image)

```python
transforms.RandomCrop(300)(img_new)
```

### 9. Случайным образом конвертирует (или не конвертирует) в оттенки серого. (Randomly convert to grayscale).

```python
transforms.RandomGrayscale()(img_new)
```

### 10. Случайным образом отражает изображение по горизонтали.(Randomly reflect image horizontally)

```python
transforms.RandomHorizontalFlip()(img_new)
```

### 11. Применяет все трасформации из переданного списка в случайном порядке. (Apply a list of the transformations in a random order).

```python
transforms.RandomOrder([transforms.Pad(padding=(100, 50, 20, 11), fill=123), transforms.Grayscale(3), \
                        transforms.RandomCrop(300)])(img_new)
```
